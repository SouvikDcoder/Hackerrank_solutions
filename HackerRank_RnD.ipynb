```
#https://www.hackerrank.com/challenges/non-divisible-subset/problem?isFullScreen=true


#!/bin/python3
import itertools as it
import math
import os
import random
import re
import sys

#
# Complete the 'nonDivisibleSubset' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER k
#  2. INTEGER_ARRAY s
#

def nonDivisibleSubset(k, s):
    # Write your code here
    remcount = [0]*k 
    #creates an list with index 0 to (k-1) which are the only possible remainders
    rval = 0 
        #this is return value initialized to 0
    for ss in s: 
        remcount[ss%k]+=1 
            #each element remainder occurence is counted in the list remcount 
    if(remcount[0]>0):
        rval += 1
        # you can only have one fully divisible element once in the sub array
        #if you have two the sum will also be fully divisible
        
    for i in range(1,(k//2)+1):
        #loop starts from 1 as if there is any fully divisible elements at all its added already
        
        if(i!=k-(k/2)): 
            rval+=max(remcount[i],remcount[k-i])
            #max operation is performed which remainder have more elements 
            #these remainers are complementary i and k-i if both are present remainder become zero so only one can be present of k and k-1 remainders 
        elif(remcount[i]>0):
            #a unique condition again occur when the remained is exacly half times k 
            #if you have such element twice remained again becomes 0 so such elements can be only added once 
            rval+= 1
                
    return rval

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    first_multiple_input = input().rstrip().split()

    n = int(first_multiple_input[0])

    k = int(first_multiple_input[1])

    s = list(map(int, input().rstrip().split()))

    result = nonDivisibleSubset(k, s)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```
## solution from the discussion 

import math

def stackable(blocks):
    bottom = math.inf
    while blocks:
        if blocks[0] >= blocks[-1] and \
           blocks[0] <= bottom:
            bottom = blocks[0]
            del blocks[0]
            continue
        if blocks[-1] >= blocks[0] and \
           blocks[-1] <= bottom:
            bottom = blocks[-1]
            del blocks[-1]
            continue
        return False
    return True

if __name__ == '__main__':
    T = int(input())
    for _ in range(T):
        n = int(input())
        blocks = list(map(int, input().split()))
        if stackable(blocks):
            print('Yes')
        else:
            print('No')
```
